```python
from soynlp.tokenizer import RegexTokenizer

tokenizer = RegexTokenizer()
```

```python
data = open('main.txt','r',encoding='utf-8')
data_write = open('write.swl','a',encoding='utf-8')

for data_text in data:
  data_word = str(tokenizer.tokenize(data_text))
  print(data_word)
  data_write.write(data_word+"\n")
```
